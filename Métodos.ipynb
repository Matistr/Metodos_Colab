# **Clase 2**

```
import pandas as pd #nos permite trabajar con bases de datos
import numpy as np #permite realizar operaciones con vectores y matrices
import matplotlib.pyplot as plt #nos permite graficar en 2D
```

```
datos={'estatura':[1.55,1.61,1.9,1.85,1.78,1.59,1.89,2.02],
       'peso':[65,89,78,101,78,65,89,100]}
df=pd.DataFrame(datos)
peso_total=df['peso'].sum()
print(f'El peso total es: {peso_total}')
```

```
print('Funcion Logaritmo')
x=np.linspace(0.1,np.e,100) #linspace(punto inicial, punto final, numero de particiones)
f=np.log(x)
plt.plot(x,f,color='purple')
plt.title('Grafica de la funcion Logaritmo')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```
#aplicar la funcion logaritmo a la columna peso:
def log(x): #definimos logaritmo
  return np.log(x)
df1=df['peso'].apply(log)
print(df1)
```

```
#graficar la funcion tangente:
x1=np.linspace(0,8*np.pi,100)
f1=np.tan(x1)
plt.plot(x1,f1,color='black')
plt.title('Grafica de la funcion tan(x)')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.show()
```

```
#graficar la funcion seno:
f2=np.sin(x1)
plt.plot(x1,f2,color='red')
plt.title('Grafica de la funcion sen(x)')
plt.xlabel('x')
plt.ylabel('sen(x)')
plt.show()
```

```
#graficar la funcion coseno:
x=np.linspace(-np.pi,np.pi,100)
f=np.cos(x)
plt.plot(x,f,color='darkgreen')
plt.title('Grafica de la funcion cos(x)')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
print('Graficar las funciones cos(x) y x^2-2x+2')
fig,ax=plt.subplots(2,1) #.subplots(n filas, n columnas)
x1=np.linspace(0.001,8*np.pi,100)
f1=np.cos(x1)
ax[0].plot(x1,f1,color='black')
ax[0].set_title('Grafica de la funcion cos(x)')

f2=x1**2-2*x1+2
ax[1].plot(x1,f2,color='black')
ax[1].set_title('Grafica de la funcion x^2-2x+2')

from google.colab import files
plt.savefig('graficas.jpg')
files.download('graficas.jpg')

plt.show()
```
